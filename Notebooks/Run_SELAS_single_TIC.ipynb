# Run SELAS for one TIC ID

This notebook imports all functions from `selas_runner.py` and runs the full SELAS pipeline for a single TIC ID.

```python
from pathlib import Path
import sys

SELAS_PATH = Path("../Selas").resolve()

# Where outputs should be saved.
DATA_ROOT = ("../Data/Selas-TIC-ids")


if str(SELAS_PATH) not in sys.path:
    sys.path.insert(0, str(SELAS_PATH))

print("SELAS_PATH:", SELAS_PATH)
print("DATA_ROOT:", DATA_ROOT)
```

```python
# Import the module, load the SELAS components, then import all functions.
import selas_runner as sr

sr.configure_environment(selas_path=SELAS_PATH)
sr.import_selas_components(selas_path=SELAS_PATH)

from selas_runner import *
```

## Choose one TIC ID

Pick any `TIC_ID` below, then run the pipeline cell.


```python
TIC_ID = "290716988"  # change this to the TIC ID you want to run
MAKE_PLOTS = True
```

```python
result = sr.run_single_tic_pipeline(
    tic_id=TIC_ID,
    data_root=DATA_ROOT,
    selas_path=SELAS_PATH,
    make_plots=MAKE_PLOTS,
    capture_log=True,
    display_dataframes=True,
    raise_on_error=True,
)

print("Run status:", result.get("status"))
print("Log file:", result.get("log_path"))
print("Output folder:", result.get("data_path"))
```
